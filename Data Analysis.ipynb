```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
df.head()
```

# gender distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

From the above chart we have analyzed that the number of females in the data is more than the number of males

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp, annot =True)
plt.title("Relationship between Parent's Education and student scores")
plt.show()
```

From the above chart we have concluded that the education of the parents have a good impact 

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp1, annot =True)
plt.title("Relationship between Parent's Marital status and student scores")
plt.show()
```

From the above chart we have concluded that there is no/negligible impact on te student score due to their martial status

# Distribution of Ethnic groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")

plt.title("Distribution of Ethnic Groups")
plt.show()
```
